## Riconoscimento di Immagini

Per andare ad allenare e testare le nostre reti neurali, utilizzeremo il database MNIST che contiene $70000$ cifre scritte a mano, ognuna rappresentata da una immagine quadrata divisa in una griglia di 28x28 pixel. Ogni pixel puo' avere una delle 256 gradazioni del grigio, dal bianco al nero, e quindi ogni punto della griglia puo' assumere un qualsiasi valore in $\{0,1,\dots,255\}$. \
Dato che andremo a catalogare le cifre assegnandogli una etichetta, da 0 a 9, questo rappresenta un tipico esempio di problema di classificazione.

Per andare a riconoscere le cifre utilizzeremo un modello di deep neural network e di convolutinal neural network sfruttando la libreria Keras di Python. \
I DNN sono costituiti da vari layer densi di neuroni e sono, all'infuori del layout scelto nello specifico task, assolutamente identici alle reti costruite nell'esercitazione 11 per effettuare operazioni di regressione.
I CNN, invece, costituiscono un approccio diverso al supervised learning: sono stati creati come architettura che sfrutta le simmetrie fisiche per guadagnare efficienza su specifici task. Essi usano filtri e layer di pooling per eccellere in compiti come l'image processing. \
In questo tipo di network la gran parte dell'ottimizzazione stocastica non sta nel trovare pesi e bias adatti, ma il tipo e il numero di filtri da utilizzare.

Per indagare come questi due tipi di network approcciano e risolvono il riconoscimento di cifre scritte a mano, l'esercitazione si compone di tre diverse parti:

- ottimizzazione di una DNN nell'[esercizio 12.1](#esercizio12.1);
- costruzione e ottimizzazione di una CNN nell'[esercizio 12.2](#esercizio12.2);
- applicazione della nostra CNN al riconoscimento di cifre scriite da noi a mano nell'[esercizio 12.3](#esercizio12.3).

## Esercizio 12.1 <a class="anchor" id="esercizio12.1"></a>

Approcciamo ai problemi di classificazione studiando come le DNN vanno a classificare le cifre da 0 a 9 fornite nel database MNIST.

Per questo esercizio il codice e' stato fornito gia' funzionante e il nostro obiettivo e' andare ad ottimizzare la rete studiando il comportamento delle loss, dell'accuratezza e le predizioni del network al variare del numero di epoche e dell'ottimizzatore scelto, tenendo come funzione costo la categorical crossentropy e come funzione di attivazione la relu. \
Mentre la funzione costo e l'ottimizzatore sono essenziali per la procedura di training, per testare le performance abbiamo scelto di osservare la accuracy, dato che affrontiamo un problema di classificazione, che e' definita come la percentuale di dati correttamente classificati.

Il network implementato e' costituito da tre layer: il primo di 400 neuroni con input shape 784, cioe' le immagini 28x28 rese vettori, il secondo e il terzo da 100 neuroni; abbiamo poi un layer di dropout per aiutare la rete a generalizzare e come output un layer costituito da 10 neuroni, come le 10 categorie da riconoscere, con funzione di attivazione la softmax per trasformare in probabilita' le componenti del vettore.

Per svolgere l'esercitazione in modo piu' efficiente abbiamo, dunque, modificato il codice per andare a plottare contemporaneamente le loss, l'accuracy e le digit predictions nel caso di diverse epoche e ottimizzatore scelto. \
Invece il numero di batch, cioe' quanti dati usiamo per calcolare un gradiente aprossimato che ci indica la direzione di spostamento per la nostra ottimizzazione, e' fissato a 32. 

```python
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True' #This is needed in my Anaconda+MacOsX installation; leave it commented.
import tensorflow as tf
from tensorflow import keras
from keras.datasets import mnist

import numpy as np
import matplotlib.pyplot as plt

seed=0
np.random.seed(seed) # fix random seed sia per il gen di numeri casuali di numpy che tf. 
#Se non lo fisso, ogni volta ne uso uno diverso invece e' sano poter riprodurre i propri risultati
tf.random.set_seed(seed)


# input image dimensions
img_rows, img_cols = 28, 28 # number of pixels 
# output
num_classes = 10 # 10 digits

from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import SGD, Adam, RMSprop, Adagrad, Adadelta, Adam, Adamax, Nadam

#creiamo il depp neural network
def create_DNN():
    # instantiate model
    model = Sequential()
    # add a dense all-to-all relu layer
    model.add(Dense(400,input_shape=(img_rows*img_cols,), activation='relu'))
    # add a dense all-to-all relu layer
    model.add(Dense(100, activation='relu'))
    # apply dropout with rate 0.5, cioe' ne spengo la meta': questo layer funziona solo nel training
    #quindi e' plausibile che la loss sul test sia minore di quella del train perche' nel test usa tutti i neuroni
    model.add(Dropout(0.5))
    #aiuta la rete a generalizzare al non fidarsi troppo di una struttura neuronale che 
    #impara in modo troppo ridigo solo quello che ha visto
    
    # soft-max layer: mi rende probabilita' le componenti di questo vettore di 10 componenti
    model.add(Dense(num_classes, activation='softmax'))
    
    return model

def compile_model(choosen_optimizer): #standard optimizer=SGD()
    # create the model
    model=create_DNN()
    # compile the model
    model.compile(loss=keras.losses.categorical_crossentropy, #la categorical tende a zero quando azzecco la label
                  optimizer=choosen_optimizer, metrics=['accuracy'])
    return model

# training parameters
batch_size = 32 
epochs = [5, 10, 15, 20] #std=5
optimizers = ['sgd', 'Adam', 'Nadam', 'Adagrad', 'Adadelta'] #adam e' sgd con momenti, nadam col momento di nesterov
#adadelta ha learning rate default=0.001

rows=len(optimizers)*len(epochs)
columns=2
j=1

for e in epochs:
    for o in optimizers:
        #se non metto la generazione dati ad ogni ciclo, fit mi da' problemi sulle dimensioni dei vettori
        # the data, split between train and test sets
        (X_train, Y_train), (X_test, Y_test) = mnist.load_data()
        #60k immagini 28x28 ognuna con la propria label (quindi ho 60k labels)


        # reshape data, it could depend on Keras backend
        X_train = X_train.reshape(X_train.shape[0], img_rows*img_cols) #li faccio diventare dei vettori 28x28 componenti
        X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)

        # cast floats to single precision
        X_train = X_train.astype('float32')#da int a float
        X_test = X_test.astype('float32')

        # cosi' ho vettori di componenti da 0 a 1
        X_train /= 255
        X_test /= 255

        # convert class vectors to binary class matrices, e.g. for use with categorical_crossentropy
        #scrivo le label di binario per poter usare la categorical crossentropy
        Y_train = keras.utils.to_categorical(Y_train, num_classes)
        Y_test = keras.utils.to_categorical(Y_test, num_classes)
        #print('X_train shape:', X_train.shape) #60k vettori da 784 componenti
        #print('Y_train shape:', Y_train.shape) #10k vettori da 784 componenti
        
        
        # create the deep neural net
        model_DNN = compile_model(o)

        # train DNN and store training info in history
        history = model_DNN.fit(X_train, Y_train,
                  batch_size=batch_size, epochs=e,
                  verbose=0, validation_data=(X_test, Y_test))

        # evaluate model e' una performance evaluation
        score = model_DNN.evaluate(X_test, Y_test, verbose=0)

        # print performance
        print()
        print('Test loss:', score[0])
        print('Test accuracy:', score[1])

        # summarize history for accuracy
        plt.figure(figsize=(11,5*rows))
        plt.subplot(rows, columns, j)
        plt.plot(history.history['accuracy'])
        plt.plot(history.history['val_accuracy'])
        plt.title('Model accuracy with: \n'+str(e)+' epochs and '+str(o)+' as the optimizer ')
        plt.ylabel('Model Accuracy')
        plt.xlabel('Epoch')
        plt.grid(True)
        plt.legend(['Train', 'Test'], loc='best')

        # summarize history for loss
        plt.subplot(rows, columns, j+1)           
        plt.title('Model loss with: \n'+str(e)+' epochs and '+str(o)+' as the optimizer ')
        plt.plot(history.history['loss'])
        plt.plot(history.history['val_loss'])
        plt.ylabel('Model Loss')
        plt.xlabel('Epoch')
        plt.grid(True)
        plt.legend(['Train', 'Test'], loc='best')
        
        plt.show()

        #print delle digits e della prediction
        predictions = model_DNN.predict(X_test)
        X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols,1)
        print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
        print ('Digits Predictions:')
        plt.figure(figsize=(15, 15)) 
        for i in range(10):    
            ax = plt.subplot(2, 10, i + 1)    
            plt.imshow(X_test[i, :, :, 0], cmap='gray')    
            plt.title("Digit: {}\nPredicted: {}".format(np.argmax(Y_test[i]), np.argmax(predictions[i])))    
            plt.axis('off') 
        plt.show()
        
        j=j+2;
```

In termini di loss, accuratezza e potere predittivo i migliori risultati si ottengono utilizzando come ottimizzatore SGD o Adagrad. Con un numero basso di epoche, pero', vediamo che le loss e l'accuratezza possono ancora diminuire, non si sono ancora stabilizzate, sintomo dell'aver utilizzato troppe poche epoche. \
Analizziamo, quindi, se aumentando il numero di epoche a 75 osserviamo una ulteriore diminuzione delle loss e aumento dell'accuratezza con i due migliori ottimizzatori studiando, inoltre, come si comporta Adam con un numero di epoche cosi' elevato.

## Esercizio 12.2 <a class="anchor" id="esercizio12.2"></a>

Fino ad ora abbiamo considerato ogni data sample del MNIST come un vettore 1D $28\times 28$. Tuttavia sappiamo che in ognuna delle cifre scritta a mano ci sono correlazioni spaziali locali tra i vari pixel e vale anche l'invarianza traslazionale; da questi simmetrie vorremo tratte vantaggio per migliorare l'accuratezza del nostro modello di classificazione. \
Reti neurali che possono imparare a riconoscere questi pattern locali sono proprio le convolutional neural network.

Andiamo, quindi, a trasformare l'architettura della nostra deep neural network in una rete convoluzionale utilizzando i layer caratteristici di questo tipo di network:
- `Conv2D` costituisce il cuore di un CNN e rappresenta i filtri e le dimensioni delle matrici che applichiamo sull'informazione;
- `MaxPooling2D` in cui l'informazione spaziale viene ridotta rimpiazzando una piccola regione dell'immagine con un solo pixel il cui valore corrisponde al massimo tra quelli presenti nella regione iniziale;
- `Dropout` per spegnere alcuni neuroni ed evitare che la rete vada ad overfittare;
- `Flatten` porta l'informazione 2D ad un vettore 1D su cui si puo' lavorare con un DNN e i suoi layer densi.

Costruiamo, percio', un semplice CNN e andiamo ad osservare come si comporta sul database del MNIST utilizzando 30 epoche.

```python
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True' #This is needed in my Anaconda+MacOsX installation; leave it commented.
import tensorflow as tf
from tensorflow import keras
from keras.datasets import mnist

import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Dropout, Conv2D, MaxPooling2D, Flatten
from keras.optimizers import SGD, Adam, RMSprop, Adagrad, Adadelta, Adam, Adamax, Nadam


seed=0
np.random.seed(seed) # fix random seed
tf.random.set_seed(seed)

# input image dimensions
img_rows, img_cols = 28, 28 # number of pixels 
# output
num_classes = 10 # 10 digits

# the data, split between train and test sets
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

# reshape data, it could depend on Keras backend
X_train = X_train.reshape(X_train.shape[0], img_rows*img_cols)
X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)

# cast floats to single precision
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# rescale data in interval [0,1]
X_train /= 255
X_test /= 255

# convert class vectors to binary class matrices, e.g. for use with categorical_crossentropy
Y_train = keras.utils.to_categorical(Y_train, num_classes)
Y_test = keras.utils.to_categorical(Y_test, num_classes)


# you will need the following for Convolutional Neural Networks
from keras.layers import Flatten, Conv2D, MaxPooling2D

# reshape data, depending on Keras backend
if keras.backend.image_data_format() == 'channels_first':
    X_train = X_train.reshape(X_train.shape[0], 1, img_rows, img_cols)
    X_test = X_test.reshape(X_test.shape[0], 1, img_rows, img_cols)
    input_shape = (1, img_rows, img_cols)
else:
    X_train = X_train.reshape(X_train.shape[0], img_rows, img_cols, 1)
    X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)
    input_shape = (img_rows, img_cols, 1)

#torno ad avere 60k immagini 28x28 e 60k label. Facciamo train su 60k e test su 10k
       
# remember to move towards a standard flat layer in the final part of your DNN
#ispirazione: https://adamharley.com/nn_vis/cnn/3d.html
def create_CNN():
    #ora ho immagine 28x28x1
    
    # instantiate model
    model = Sequential()
    # add first convolutional layer with 10 filters (dimensionality of output space cioe' 5x5)
    #10 filtri applico la convoluzione con una mat 5x5
    model.add(Conv2D(10, kernel_size=(5, 5), activation='relu', input_shape=input_shape))
    #qui non volevo dropout perche sono pochi
    model.add(MaxPooling2D(pool_size=(2,2)))
    
    model.add(Conv2D(25, kernel_size=(5, 5), activation='relu'))
    model.add(Dropout(0.3))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    
    model.add(Flatten())
    model.add(Dense(200, activation='relu'))
    model.add(Dense(150, activation='relu'))
    
    model.add(Dense(10, activation='softmax')) 
    #we need a soft-max layer with num_classes=10 possible outputs as output layer
    
    # compile the model
    model.compile(loss=keras.losses.categorical_crossentropy, optimizer='SGD', metrics=['acc'])
    return model


# training parameters
batch_size = 32
epochs = 30 #forse 15, dipende da 12.1

# create the deep conv net
model_CNN=create_CNN()

# train CNN
history = model_CNN.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=0,
          validation_data=(X_test, Y_test))

# evaluate model
score = model_CNN.evaluate(X_test, Y_test, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# look into training history

plt.figure(figsize=(11,5))

# summarize history for accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Model Accuracyvs Epochs')
plt.xlabel('Epoch')
plt.grid(True)
plt.legend(['Train', 'Test'], loc='best')

# summarize history for loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss vs Epochs')
plt.ylabel('Model Loss')
plt.xlabel('Epoch')
plt.grid(True)
plt.legend(['Train', 'Test'], loc='best')

plt.suptitle("Convolutional Neural Net")
plt.show()

predictions = model_CNN.predict(X_test)
X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols,1)
print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
print ('Digits Predictions:')
plt.figure(figsize=(15, 15)) 
for i in range(10):    
    ax = plt.subplot(2, 10, i + 1)    
    plt.imshow(X_test[i, :, :, 0], cmap='gray')    
    plt.title("Digit: {}\nPredicted:{}".format(np.argmax(Y_test[i]), np.argmax(predictions[i])))    
    plt.axis('off') 
plt.show()
```

```python
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
import tensorflow as tf
from tensorflow import keras
from keras.datasets import mnist

import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Dropout, Conv2D, MaxPooling2D, Flatten
from keras.optimizers import SGD, Adam, RMSprop, Adagrad, Adadelta, Adam, Adamax, Nadam


seed=0
np.random.seed(seed) # fix random seed
tf.random.set_seed(seed)

# input image dimensions
img_rows, img_cols = 28, 28 # number of pixels 
# output
num_classes = 10 # 10 digits

# the data, split between train and test sets
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

# reshape data, it could depend on Keras backend
X_train = X_train.reshape(X_train.shape[0], img_rows*img_cols)
X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)

# cast floats to single precision
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# rescale data in interval [0,1]
X_train /= 255
X_test /= 255

# convert class vectors to binary class matrices, e.g. for use with categorical_crossentropy
Y_train = keras.utils.to_categorical(Y_train, num_classes)
Y_test = keras.utils.to_categorical(Y_test, num_classes)


# you will need the following for Convolutional Neural Networks
from keras.layers import Flatten, Conv2D, MaxPooling2D

# reshape data, depending on Keras backend
if keras.backend.image_data_format() == 'channels_first':
    X_train = X_train.reshape(X_train.shape[0], 1, img_rows, img_cols)
    X_test = X_test.reshape(X_test.shape[0], 1, img_rows, img_cols)
    input_shape = (1, img_rows, img_cols)
else:
    X_train = X_train.reshape(X_train.shape[0], img_rows, img_cols, 1)
    X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)
    input_shape = (img_rows, img_cols, 1)

    
# remember to move towards a standard flat layer in the final part of your DNN
#FOR EXAMPLE SOME: Dropout, 2D pooling, 2D convolutional etc. ... https://adamharley.com/nn_vis/cnn/3d.html
#ispirazione: https://adamharley.com/nn_vis/cnn/3d.html
def create_CNN():
    #ora ho immagine 28x28x1
    
    # instantiate model
    model = Sequential()
    # add first convolutional layer with 10 filters (dimensionality of output space cioe' 5x5)
    model.add(Conv2D(10, kernel_size=(5, 5), activation='relu', input_shape=input_shape)) #primi 10 filtri
    #qui non volevo dropout perche sono pochi
    model.add(MaxPooling2D(pool_size=(2,2)))
    
    model.add(Conv2D(25, kernel_size=(5, 5), activation='relu'))
    model.add(Dropout(0.3))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    
    model.add(Flatten())
    model.add(Dense(200, activation='relu'))
    model.add(Dense(150, activation='relu'))
    
    #droout da qualche oarte per impedirgli di overfittare?????? 
    model.add(Dense(10, activation='softmax')) 
    #we need a soft-max layer with num_classes=10 possible outputs as output layer
    
    # compile the model
    model.compile(loss=keras.losses.categorical_crossentropy, optimizer='SGD', metrics=['acc'])
    return model


# training parameters
batch_size = 32
epochs = 20 #forse 15, dipende da 12.1

# create the deep conv net
model_CNN=create_CNN()

# train CNN
history = model_CNN.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=0,
          validation_data=(X_test, Y_test))

# evaluate model
score = model_CNN.evaluate(X_test, Y_test, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# look into training history

plt.figure(figsize=(11,5))

# summarize history for accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Model Accuracyvs Epochs')
plt.xlabel('Epoch')
plt.grid(True)
plt.legend(['Train', 'Test'], loc='best')

# summarize history for loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss vs Epochs')
plt.ylabel('Model Loss')
plt.xlabel('Epoch')
plt.grid(True)
plt.legend(['Train', 'Test'], loc='best')

plt.suptitle("Convolutional Neural Net")
plt.show()

predictions = model_CNN.predict(X_test)
X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols,1)
print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
print ('Digits Predictions:')
plt.figure(figsize=(15, 15)) 
for i in range(10):    
    ax = plt.subplot(2, 10, i + 1)    
    plt.imshow(X_test[i, :, :, 0], cmap='gray')    
    plt.title("Digit: {}\nPredicted:{}".format(np.argmax(Y_test[i]), np.argmax(predictions[i])))    
    plt.axis('off') 
plt.show()
```
